```python
"""
날짜 : 2022/04/26
이름 : 양용민
내용 : 파이썬 Seaborn 실습하기

Seaborn
 - matplot 라이브러리를 기반으로 만들어진 라이브러리
 - matplorlib 보다 코드가 쉽고 간결한 고급 시각화 지원
"""
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# 막대그래프
plt.title('CountPlot')
sns.countplot(data=tips, x='day' )
plt.show()
```

```python
# 히스토그램
plt.title('Histogram')
sns.distplot(tips['total_bill'])
plt.show()
```

```python
# 산점도
plt.title('Scatter')
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# 박스상자
plt.title('BoxPlot')
sns.boxplot(data=tips, x='time', y='total_bill')
plt.show()
```

```python
# 통합 시각화
sns.pairplot(tips)
plt.show()
```
