```python
# 버섯 데이터 머신러닝
import pandas as pd
from sklearn import svm, metrics, model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# 데이터 불러오기
mr = pd.read_csv('/content/drive/MyDrive/파이썬 데이터 과학 실습/file/mushroom.csv', header=None)
mr
```

```python
# 데이터 정규화(문자 데이터를 숫자 데이터로 변환)
data = []
label = []

for row_index, row in mr.iterrows():
  label.append(row.iloc[0])
  row_data = []
  for v in row.iloc[1:]:
    row_data.append(ord(v))

  data.append(row_data)
```

```python
# 학습데이터, 테스트데이터
train_data, test_data, train_label, test_label = train_test_split(data, label, train_size = 0.8)
```

```python
# 모델 학습하기
model = svm.SVC()
model.fit(data, label)
```

```python
# 모델 테스트
result = model.predict(test_data)
result
```

```python
# 모델 성능평가
score = metrics.accuracy_score(result, test_label)
print(score)
```
